```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime
```

```python
df = pd.DataFrame(pd.read_csv('./behavioural - EX.csv', header=0))

# Filter a video to process

df = df.loc[lambda df: df['Video'] == 'Example3']         #---------modify

#df['At_video'].astype('int')

df = df.sort_values(['At_video'])
df = df.reset_index(drop=True)

rows = df.iloc[:,0].size

print(df)
```

```python
# Set a threshold number to control the limitition of interval
theta = 4
```

```python
df.info()
```

```python
i = 0
k = 1
commentid_list = [str(df['CommentId'][i])]
intervals = [k]


for i in range(rows - 1):
    start = df['At_video'][i]
    end = df['At_video'][i + 1]
    gap = end - start
    #print(gap)
    
    if gap < theta:
        label = k
        #print(label)
        intervals.append(label)
        commentid_list.append(str(df['CommentId'][i + 1]))
        i = i + 1
    else:
        label = k + 1
        #print(label)
        intervals.append(label)
        commentid_list.append(str(df['CommentId'][i + 1]))
        i = i + 1
        k = k + 1
#print(commentid_list)
results = pd.Series(intervals, index = commentid_list)
print(results)
```

```python
# save the prediction results
j = 0
record_count = 0
my_file = open('./interval_output/results_example3_i4.txt','w')  #---------modify
for _ in results:
    my_file.writelines(commentid_list[j])
    my_file.writelines(',')
    my_file.writelines(str(_))
    my_file.writelines('\n')
    record_count += 1
    j = j + 1
my_file.close()

print("The number of comments: ", record_count)
print("Finished")
```
